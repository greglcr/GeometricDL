```python
import subprocess
import os

def compile_latex_in_notebook(file_path):
    try:
        # Get the directory and filename
        directory = os.path.dirname(file_path)
        file_name = os.path.basename(file_path)
        base_name = os.path.splitext(file_name)[0]
        
        # Function to run a subprocess command
        def run_command(command, cwd):
            result = subprocess.run(
                command,
                cwd=cwd,
                text=True,
                capture_output=True
            )
            print(f"Running: {' '.join(command)}")
            if result.returncode != 0:
                print(f"Error:\n{result.stderr}")
                raise subprocess.CalledProcessError(result.returncode, command)
            print(result.stdout)

        # Run pdflatex -> bibtex -> pdflatex -> pdflatex
        run_command(["pdflatex", file_name], directory)
        run_command(["bibtex", base_name], directory)
        run_command(["pdflatex", file_name], directory)
        run_command(["pdflatex", file_name], directory)

        print(f"\nPDF successfully compiled at: {os.path.join(directory, file_name.replace('.tex', '.pdf'))}")
    except subprocess.CalledProcessError as e:
        print("An error occurred during compilation.")
    except FileNotFoundError:
        print("pdflatex or bibtex not found. Ensure they are installed and added to your system PATH.")

# Specify the file path
file_path = r"C:\Users\ns243\Documents\Academic\AI Master\M2\Courses\T2\GDA\Project\GeometricDL-main\Report\Geometric Data Analysis-Sahebalzamany-Rouvroy-Le Corre.tex"
compile_latex_in_notebook(file_path)

```
